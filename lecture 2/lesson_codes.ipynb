##### Bubble Sort

1. Go through the array, one value at a time.
2. For each value, compare the value with the next value.
3. If the value is higher than the next one, swap the values so that the highest value comes last.
4. Go through the array as many times as there are values in the array.

```python
bb_sort_array = [1, 3, 23, 11, 8, 1]

size = len(bb_sort_array)

for x in range(size-1):
    for y in range(size-x-1):

        if bb_sort_array[y] > bb_sort_array[y+1]:
            bb_sort_array[y], bb_sort_array[y +
                                            1] = bb_sort_array[y+1], bb_sort_array[y]


print(bb_sort_array)
```

##### Selection Sort

1. An array with values to sort.
2. An inner loop that goes through the array, finds the lowest value, and moves it to the front of the array. This loop must loop through one less value each time it runs.
3. An outer loop that controls how many times the inner loop must run. For an array with n values, this outer loop must run n−1 times.

```python
sel_sort_array = [34, 22, 1, 33, 12, 3, 5, 66, 12,  11]

size = len(sel_sort_array)

for x in range(size - 1):
    index =  x  # index of min value
    for y in range(x+1, size):

        if sel_sort_array[y] < sel_sort_array[index]:
            index  =  y # index of current min value
    
    min_value = sel_sort_array.pop(index)
    sel_sort_array.insert(x, min_value)


print(sel_sort_array)
```

```python
# making it better with swaping instead of remove and insert
sel_sort_array = [34, 22, 1, 33, 12, 3, 5, 66, 12,  11]

size = len(sel_sort_array)

for x in range(size - 1):
    index =  x  # index of min value
    for y in range(x+1, size):

        if sel_sort_array[y] < sel_sort_array[index]:
            index  =  y # index of current min value
    
    sel_sort_array[x], sel_sort_array[index] = sel_sort_array[index], sel_sort_array[x]

print(sel_sort_array)
```

#### Insertion sort

1. An array with values to sort.
2. An outer loop that picks a value to be sorted. For an array with n values, this outer loop skips the first value, and must run n−1times.
3. An inner loop that goes through the sorted part of the array, to find where to insert the value. If the value to be sorted is at index i, the sorted part of the array starts at index 0
and ends at index i−1.

```python
inse_sort_array = [34, 22, 1, 33, 12, 3, 5, 66, 12,  11]

size = len(sel_sort_array)

for x in range(1, size):
    index = x # insertion index
    value  =  inse_sort_array[x]

    for y in range(x-1, -1, -1):
        if inse_sort_array[y] > value:
            inse_sort_array[y+1] = inse_sort_array[y]
            index = y # current insertion index

        else:
            break
    inse_sort_array[index] = value

print(inse_sort_array)
```

#### Quicksort

The algorithm chooses a 'pivot' element then moves lower values to left and higher values to the right

1. Choose a value in the array to be the pivot element.
2. Order the rest of the array so that lower values than the pivot element are on the left, and higher values are on the right.
3. Swap the pivot element with the first element of the higher values so that the pivot element lands in between the lower and higher values.
4. Do the same operations (recursively) for the sub-arrays on the left and right side of the pivot element.



 A partition or split method that receives a sub-array, moves values around, swaps the pivot element into the sub-array and returns the index where the next split in sub-arrays happens.

```python
def split(list, low, high):
    pivot = list[high]
    x = low - 1

    for y in range(low, high):
        if list[y] <= pivot:
            x += 1
            list[x], list[y] = list[y], list[x]

    list[x+1], list[high] = list[high], list[x+1]
    return x+1


def quicksort(array, low=0, high=None):
    if high is None:
        high = len(array) - 1

    if low < high:
        pivot_index = split(array, low, high)
        quicksort(array, low, pivot_index-1)
        quicksort(array, pivot_index+1, high)
```

```python
quick_sort_array = [34, 22, 1, 93, 12, 3, 5, 66, 12,  11]
quicksort(quick_sort_array)

print(quick_sort_array)
```

##### Merge Sort

1. An array with values that needs to be sorted.
2. A function that takes an array, splits it in two, and calls itself with each half of that array so that the arrays are split again and again recursively, until a sub-array only consist of one value.
3. Another function that merges the sub-arrays back together in a sorted way.

```python
def merge_sort(list):
    if len(list) <= 1:
        return list

    mid = len(list) // 2
    leftHalf = list[:mid]
    rightHalf = list[mid:]

    sortedLeft = merge_sort(leftHalf)
    sortedRight = merge_sort(rightHalf)

    return merge(sortedLeft, sortedRight)


def merge(left, right):
    result = []
    x = y = 0

    while x < len(left) and y < len(right):
        if left[x] < right[y]:
            result.append(left[x])
            x += 1
        else:
            result.append(right[y])
            y += 1

    result.extend(left[x:])
    result.extend(right[y:])

    return result
```

```python
merge_sort_array = [2, 4, -1, 3, 12.6, 90, -40, 1, 12, -99, 20]

sorted_list =  merge_sort(merge_sort_array)
print(sorted_list)
```

#### Graph Algorithms

Example of Graph representation Adjacency matrix

```python
class Graph:
    def __init__(self, V, directed=0):
        self.V = V
        self.directed = directed
        self.Adj = [[0 for _ in range(V)] for _ in range(V)]  # Initialize a 2D boolean matrix

        for i in range(V):
            self.Adj[i][i] = 0

    
    def _print(self):
        print(self.Adj)

    def add_edge(self, u, v):
        self.Adj[u][v] = 1
        if not self.directed:
            self.Adj[v][u] = 1

    def display(self):
        for row in self.Adj:
            print(row)
```

```python
graph = Graph(4, directed=0)
graph.add_edge(0, 3)
graph.display()
```

Example of Graph representation Adjency List

```python
class ListNode:
    def __init__(self, vertex):
        self.vertexNumber = vertex
        self.next = None


class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.adj = [None] * vertices  # Initialize adjacency list

        for i in range(vertices):
            self.adj[i] = ListNode(i)


def adj_list_of_graph():
    try:

        V = int(input("Number of Vertices: "))
        E = int(input("Number of Edges: "))

        graph = Graph(V)

        for i in range(E):
            x, y = map(int, input(f"Reading Edge {  i + 1} (e.g., 'x y'): ").split())

            temp = ListNode(y)
            temp.next = graph.adj[x].next
            graph.adj[x].next = temp

            temp = ListNode(x)
            temp.next = graph.adj[y].next
            graph.adj[y].next = temp

        return graph
    except Exception as e:
        print(f"Error: {e}")
        return None


def print_adj_list(graph):
    if not graph:
        print("Graph is empty.")
        return

    print("Adjacency List Representation:")
    for i in range(graph.V):
        print(f"Vertex {i}:", end="")
        temp = graph.adj[i].next
        while temp:
            print(f" -> {temp.vertexNumber}", end="")
            temp = temp.next
        print()


graph = adj_list_of_graph()
if graph:
    print_adj_list(graph)
```
